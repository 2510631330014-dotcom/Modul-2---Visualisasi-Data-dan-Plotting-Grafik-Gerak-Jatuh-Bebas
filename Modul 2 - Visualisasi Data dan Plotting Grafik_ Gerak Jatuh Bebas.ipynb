```python
import matplotlib.pyplot as plt

g = 9.8
h0 = 100
v0 = 0
t = list(range(11))  # 0–10 detik

s = [h0 - 0.5 * g * (ti ** 2) for ti in t]
v = [v0 + g * ti for ti in t]
a = [g for _ in t]

plt.plot(t, s, marker='o', label="Posisi (m)")
plt.plot(t, v, marker='s', label="Kecepatan (m/s)")
plt.plot(t, a, marker='^', label="Percepatan (m/s²)")
plt.title("Kasus 1 – Standar Jatuh Bebas")
plt.xlabel("Waktu (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Maksimum Posisi:", max(s), "meter")
print("Maksimum Kecepatan:", max(v), "m/s")
```
